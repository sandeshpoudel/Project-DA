```python
import pandas as pd  
import matplotlib.pyplot as plt  
```

#Load merged data

```python
df=pd.read_csv("./merged_with_email.csv")
```

#Lets generate some very basic insight

```python
top3 = df.sort_values('id', ascending=True).head(3)
avg_rating = df['value'].mean()
```

##Lets plot

```python
plt.bar(df['title'],df['value'], color = 'skyblue')
plt.xticks(rotation=90)
plt.title("IMDB Rating for top 25")
plt.savefig("rating_plot.png", bbox_inches='tight')
```

```python
html_report = f"""
<h1>Data Analysis Report</h1>
<h3>Top 3 Movies</h3>
{top3.to_html()}
<h3> Average value:{avg_rating:.1f}/10</h3>
<img src='rating_plot.png' width ='600'>
"""
with open ("report.html","w") as f:
    f.write(html_report)
```
